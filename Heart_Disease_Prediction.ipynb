```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
```

```python
heart_data = pd.read_csv("/content/drive/MyDrive/heart.csv")
```

```python
heart_data.head()
```

```python
sns.distplot(heart_data['chol'],color="red")
```

```python
x = heart_data.drop(columns = 'target', axis = 1)
y = heart_data['target']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 2,stratify = y,test_size=0.2)
```

```python
model = LogisticRegression(max_iter = 1000)
```

```python
model.fit(x_train.values,y_train.values)
```

```python
train_accuracy = model.predict(x_train.values)
predict_train = accuracy_score(train_accuracy,y_train)
```

```python
print(predict_train)
```

```python
test_accuracy = model.predict(x_test.values)
predict_test = accuracy_score(test_accuracy,y_test)
```

```python
print(predict_test)
```
